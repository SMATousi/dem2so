```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from scipy.ndimage import binary_dilation
from matplotlib.colors import LinearSegmentedColormap

# Open the stream order map
stream_order_map_path = '/home/macula/SMATousi/Gullies/SO_Paper/data_to_send/Ellis/SO_01_DL.tif'
# stream_order_map_path = '/home/macula/SMATousi/Gullies/ground_truth/google_api/training_process/gully_detection/pos_test_1/stream_order/KS1_stream_tile_0.tif'

with rasterio.open(stream_order_map_path) as src:
    stream_order_map = src.read(1)
    transform = src.transform

# Define the stream orders to highlight
stream_orders_to_highlight = list(range(2, 9))
colors = ['blue', 'green', 'yellow', 'orange', 'red', 'purple', 'brown']

# Create a blank canvas for the highlighted map
highlighted_map = np.zeros_like(stream_order_map, dtype=np.uint8)

# Function to create buffer (dilation) around stream orders
def create_buffer(mask, iterations=3):
    return binary_dilation(mask, iterations=iterations).astype(np.uint8)

# Apply buffers and colors
for stream_order, color in zip(stream_orders_to_highlight, colors):
    mask = (stream_order_map == stream_order).astype(np.uint8)
    buffered_mask = create_buffer(mask)
    highlighted_map[buffered_mask == 1] = stream_order

# Define a colormap for the highlighted stream orders
cmap = ListedColormap(['black'] + colors)

# Plot the original and highlighted maps
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
ax1.set_title('Original Stream Order Map')
ax1.imshow(stream_order_map, cmap='gray')
ax2.set_title('Highlighted Stream Order Map')
highlighted_plot = ax2.imshow(highlighted_map, cmap=cmap, interpolation='none')
plt.colorbar(highlighted_plot, ax=ax2, ticks=stream_orders_to_highlight, label='Stream Order')
plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

tile_number = 7
# Open the stream order map
# stream_order_map_path = '/home/macula/SMATousi/Gullies/SO_Paper/data_to_send/Ellis/SO_01_DL.tif'
stream_order_map_path = f'/home/macula/SMATousi/Gullies/ground_truth/google_api/training_process/gully_detection/pos_test_1/stream_order/KS1_stream_tile_{tile_number}.tif'

with rasterio.open(stream_order_map_path) as src:
    stream_order_map = src.read(1)
    transform = src.transform
    
# Define the stream orders to highlight
stream_orders_to_highlight = list(range(2, 9))
colors = ['blue', 'green', 'yellow', 'orange', 'red', 'purple', 'brown']

# Create a blank canvas for the highlighted map
highlighted_map = np.zeros_like(stream_order_map, dtype=np.uint8)

# Function to create buffer (dilation) around stream orders
def create_buffer(mask, iterations=3):
    return binary_dilation(mask, iterations=iterations).astype(np.uint8)

# Apply buffers and colors
for stream_order, color in zip(stream_orders_to_highlight, colors):
    mask = (stream_order_map == stream_order).astype(np.uint8)
    buffered_mask = create_buffer(mask)
    highlighted_map[buffered_mask == 1] = stream_order

# Define a colormap for the highlighted stream orders
cmap = ListedColormap(['black'] + colors)

# Open the gully lines map
gully_lines_path = f'/home/macula/SMATousi/Gullies/ground_truth/google_api/training_process/gully_detection/pos_images_1/ground_truth/ground_truth_tile_{tile_number}.tif'
with rasterio.open(gully_lines_path) as src:
    gully_lines = src.read(1)

# Define the risk threshold (assuming higher values indicate higher risk)
risk_threshold = 3  # Adjust based on your risk map's scale

# Create a binary mask for high-risk areas
high_risk_mask = (highlighted_map >= risk_threshold).astype(np.uint8)

# Ensure the gully lines map has the same shape as the risk map
assert highlighted_map.shape == gully_lines.shape, "The risk map and gully lines map must have the same dimensions."

# Calculate the overlap between gully lines and high-risk areas
overlap = high_risk_mask * gully_lines
overlap_percentage = (np.sum(overlap) / np.sum(gully_lines)) * 100

# Plot the results
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
ax1.set_title('Risk Map')
ax1.imshow(highlighted_map, cmap='gray')

ax2.set_title('Gully Lines Map')
ax2.imshow(gully_lines, cmap='Blues')

ax3.set_title('Overlay of Gully Lines on High-Risk Areas')
ax3.imshow(high_risk_mask, cmap='Reds', alpha=0.5)
ax3.imshow(gully_lines, cmap='Blues', alpha=0.5)
plt.show()

print(f'Overlap Percentage: {overlap_percentage:.2f}%')
```
